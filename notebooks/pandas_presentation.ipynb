##Exercice 7 - Bloc 7

###Q1- Connection de Pandas à une base de données SQLite

```python
import sqlite3
import pandas as pd

# Créer/ouvrir la base de données
conn = sqlite3.connect('p8_devdata.db')
print('Connexion établie !')
```

###Q2- Lecture d'une table SQL dans un DataFrame

```python
apprenant.to_sql('apprenants', conn, if_exists='replace', index=False)

# Puis on le relit depuis SQL
apprenant_sql = pd.read_sql('SELECT * FROM apprenants', conn)
print(apprenant_sql.head())
```

###Q3- Affichage des données recupérées

```python
print(f'Lignes lues depuis SQL : {len(apprenant_sql)}')
print(apprenant_sql.dtypes)
```

###Q4- Filtrage des données avec pandas après lecture

```python
# Option 1 : filtrer directement dans la requête SQL
admis_sql = pd.read_sql(
    "SELECT * FROM apprenants WHERE Statut = 'Admis'",
    conn
)

# Option 2 : lire tout puis filtrer avec Pandas
admis_pd = apprenant_sql[apprenant_sql['Statut'] == 'Admis']

print(f'Admis (SQL)    : {len(admis_sql)}')
print(f'Admis (Pandas) : {len(admis_pd)}')
```

###Q5- Sauvegarde d'un dataframe dans une base de données

```python
# Créer un résumé par filière et le sauvegarder
resume = apprenant.groupby('Filiere').agg(
    Effectif=('Nom','count'),
    Moy_Python=('Note_Python','mean'),
    Moy_SQL=('Note_SQL','mean'),
).round(2).reset_index()

resume.to_sql('resume_filieres', conn, if_exists='replace', index=False)
print('Table resume_filieres créée !')
```

###Q6- Comparaison d'une opération SQL et Pandas

```python
# === SQL ===
requete_sql = '''
    SELECT Filiere,
           COUNT(*) AS Effectif,
           AVG(Note_Python) AS Moy_Python
    FROM apprenants
    GROUP BY Filiere
    ORDER BY Moy_Python DESC
'''
print(pd.read_sql(requete_sql, conn))

# === PANDAS équivalent ===
print(
    apprenant.groupby('Filiere')
      .agg(Effectif=('Nom','count'), Moy_Python=('Note_Python','mean'))
      .round(2)
      .sort_values('Moy_Python', ascending=False)
)
```
